```python
import pandas as pd
# Load the CSV file into a pandas DataFrame
df = pd.read_csv(r'D:\mongodb to pandas datafram\test_jbm.csv')

df
```

```python
# select two columns
newdf = df[['ERFURT_01-Active_Energy', 'timestamp']]
newdf
```

```python
# Remove Rows With Null Values
newdf = newdf.dropna()
newdf
```

```python
newdf['timestamp'] = pd.to_datetime(newdf['timestamp'])
print(newdf.to_string())
```

```python
# group the dataframe by minute and select the first row of each group
newdf = newdf.groupby(pd.Grouper(key='timestamp', freq='1min')).first().reset_index()
newdf
```

```python
# Remove Rows With Null Values
newdf = newdf.dropna()
newdf
```

```python
#Difference between rows
diff_df = newdf.diff()
diff_df
```
